There are outliers in certain columns.We have to remove the outliers.

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
threshold=3
np.where(z>3)
```

Six Outliers found and removed from the dataset

```python
Data_New=df[(z<3).all(axis=1)]
```

```python
Data_New
```

```python
x=Data_New.iloc[:,1:8]
y=Data_New['Happiness Score']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, random_state=42, test_size=0.33)
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
lm.score(x_train,y_train)
pred=lm.predict(x_test)
print('Error LM:')
print("Mean absolute error:",mean_absolute_error(y_test,pred))
print('Mean squared error:',mean_squared_error(y_test,pred))
print('Root Mean square error:',np.sqrt(mean_squared_error(y_test,pred)))
print('r2_score:',r2_score(y_test,pred))
```

```python
t=np.array([1.30232,1.40223,0.94784,0.62877,0.14145,0.43630,2.70201])
t=t.reshape(1,-1)        #reshaped into 2D data,where -1:Many columns,1:number of rows
t.shape
```

```python
lm.predict(t)
```

```python
svr = SVR()
dtr= DecisionTreeRegressor()
rfr = RandomForestRegressor()
knr = KNeighborsRegressor()
```

```python

dtr.fit(x_train,y_train)
dtr.score(x_train,y_train)
pred1=dtr.predict(x_test)
print('Error DTR:')
print("Mean absolute error:",mean_absolute_error(y_test,pred1))
print('Mean squared error:',mean_squared_error(y_test,pred1))
print('Root Mean square error:',np.sqrt(mean_squared_error(y_test,pred1)))
print(r2_score(y_test,pred1))
```

```python

knr.fit(x_train,y_train)
knr.score(x_train,y_train)
pred2=knr.predict(x_test)
print('Error KNR:')
print("Mean absolute error:",mean_absolute_error(y_test,pred2))
print('Mean squared error:',mean_squared_error(y_test,pred2))
print('Root Mean square error:',np.sqrt(mean_squared_error(y_test,pred2)))
print(r2_score(y_test,pred2))
```

```python

svr.fit(x_train,y_train)
svr.score(x_train,y_train)
pred3=svr.predict(x_test)
print('Error SVR:')
print("Mean absolute error:",mean_absolute_error(y_test,pred3))
print('Mean squared error:',mean_squared_error(y_test,pred3))
print('Root Mean square error:',np.sqrt(mean_squared_error(y_test,pred3)))
print(r2_score(y_test,pred3))
```

```python

rfr.fit(x_train,y_train)
rfr.score(x_train,y_train)
pred4=rfr.predict(x_test)
print('Error RFR:')
print("Mean absolute error:",mean_absolute_error(y_test,pred4))
print('Mean squared error:',mean_squared_error(y_test,pred4))
print('Root Mean square error:',np.sqrt(mean_squared_error(y_test,pred4)))
print(r2_score(y_test,pred4))
```

```python
print(lm.predict(t))
print(dtr.predict(t))
print(knr.predict(t))
print(svr.predict(t))
print(rfr.predict(t))
```
