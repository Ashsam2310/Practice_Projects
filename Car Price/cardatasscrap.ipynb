```python
for i in nam_str:
    year.append(i.text.split()[-1])
    brand.append(i.text.split()[0])
    if i.text.split()[0]=='Maruti':
        model.append(i.text.split()[2])
        b=i.text.split()[3:-1]
        if len(b)==8:
            variant.append(b[0]+" "+b[1]+" "+b[2]+" "+b[3]+" "+b[4]+" "+b[5]+" "+b[6]+" "+b[7])
        if len(b)==7:
            variant.append(b[0]+" "+b[1]+" "+b[2]+" "+b[3]+" "+b[4]+" "+b[5]+" "+b[6])
        if len(b)==6:
            variant.append(b[0]+" "+b[1]+" "+b[2]+" "+b[3]+" "+b[4]+" "+b[5])
        if len(b)==5:
            variant.append(b[0]+" "+b[1]+" "+b[2]+" "+b[3]+" "+b[4])
        if len(b)==4:
            variant.append(b[1]+" "+b[2]+" "+b[3]+" "+b[0])
        if len(b)==3:
            variant.append(b[0]+" "+b[1]+" "+b[2])
        if len(b)==2:
            variant.append(b[0]+" "+b[1])
        elif len(b)==1:
            variant.append(b[0])
    else:
        model.append(i.text.split()[1])
        b=i.text.split()[2:-1]
        if len(b)==8:
            variant.append(b[0]+" "+b[1]+" "+b[2]+" "+b[3]+" "+b[4]+" "+b[5]+" "+b[6]+" "+b[7])
        if len(b)==7:
            variant.append(b[0]+" "+b[1]+" "+b[2]+" "+b[3]+" "+b[4]+" "+b[5]+" "+b[6])
        if len(b)==6:
            variant.append(b[0]+" "+b[1]+" "+b[2]+" "+b[3]+" "+b[4]+" "+b[5])
        if len(b)==5:
            variant.append(b[0]+" "+b[1]+" "+b[2]+" "+b[3]+" "+b[4])
        if len(b)==4:
            variant.append(b[0]+" "+b[1]+" "+b[2]+" "+b[3])
        if len(b)==3:
            variant.append(b[0]+" "+b[1]+" "+b[2])
        if len(b)==2:
            variant.append(b[0]+" "+b[1])
        elif len(b)==1:
            variant.append(b[0])
```
