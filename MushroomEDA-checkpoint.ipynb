```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
DF=df1.apply(LabelEncoder().fit_transform)
DF.head()
```

Using Label Encoder to convert all the ordinal values to numerical.

```python
plt.figure(figsize=(16,12))
sns.heatmap(DF.corr(),annot=True)
```

```python
x=DF.iloc[:,1:22]
```

```python
x.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x1 = sc.fit_transform(x)
```

```python
print(x1)
```

```python
#using principal component analysis
#Even though the number of variables is not too high, I would still like to use PCA to see which variables describe the maximum variance in data
from sklearn.decomposition import PCA
pca = PCA()
x_pca = pca.fit_transform(x1)
```

```python
new_pca = PCA(n_components=17)
```

```python
x_new = new_pca.fit_transform(x1)
```

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=2)
k_means.fit_predict(x_new )
colors = ['r','g']
for i in range(len(x_new)):
    plt.scatter(x_new[i][0], x_new[i][1], c=colors[k_means.labels_[i]], s=10)
plt.show()
```

```python
y=df1.iloc[:,0]
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier

DTC=DecisionTreeClassifier(random_state=0,max_depth=5)
DTC.fit(x_train,y_train)
```

```python
DTC.score(x_train,y_train)
```

```python
pred=DTC.predict(x_test)
```

```python
accuracy_score(y_test,pred)
```

```python
cls=classification_report(y_test,pred)
print(cls)
```

```python
RFC=RandomForestClassifier()
RFC.fit(x_train,y_train)
pred1=RFC.predict(x_test)
cls1=classification_report(y_test,pred1)
print(cls1)
```

```python
SV=SVC()
SV.fit(x_train,y_train)
pred2=SV.predict(x_test)
cls2=classification_report(y_test,pred2)
print(cls2)
```

```python
KNN= KNeighborsClassifier( )  
KNN.fit(x_train, y_train)
pred3=KNN.predict(x_test)
cls3=classification_report(y_test,pred3)
print(cls3)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter={'max_depth':np.arange(2,10),
          'criterion':['gini','entropy'],
          'splitter':['best','random'],
          'max_features':['auto','sqrt','log2']}
```

```python
GCV=GridSearchCV(DecisionTreeClassifier(),parameter,cv=5)
```

```python
final_mod=DecisionTreeClassifier(criterion='entropy',max_depth=9,max_features='sqrt',splitter='best')
final_mod.fit(x_train,y_train)
pred=final_mod.predict(x_test)
acc=accuracy_score(y_test,pred)
print(acc)
```
